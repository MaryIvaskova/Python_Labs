```python
def is_number(x):
    return isinstance(x, (int, float)) and not isinstance(x, bool)

def temp_status(temp):
    if not is_number(temp):
        return "Помилка даних"
    if temp < 5:
        return "Надто холодно"
    elif temp > 25:
        return "Надто жарко"
    else:
        return "Норма"

def category_status(cat):
    s = (cat or "").strip().lower()
    match s:
        case "antibiotic":
            return "Рецептурний препарат"
        case "vitamin":
            return "Вільний продаж"
        case "vaccine":
            return "Потребує спецзберігання"
        case _:
            return "Невідома категорія"

# список препаратів
meds = []

def print_meds(meds):
    if not meds:
        print("\nСписок препаратів порожній\n")
        return

    results = []
    for d in meds:
        name = d.get("назва", "")
        qty = d.get("кількість", "")
        cat = d.get("категорія", "")
        temp = d.get("температура", "")
        if not is_number(qty) or not is_number(temp):
            results.append((name, "Помилка даних", "Помилка даних"))
        else:
            results.append((name, category_status(cat), temp_status(temp)))

    headers = ("Назва", "Категорія", "Температура")
    w1 = max(len(headers[0]), *(len(r[0]) for r in results))
    w2 = max(len(headers[1]), *(len(r[1]) for r in results))
    w3 = max(len(headers[2]), *(len(r[2]) for r in results))

    line = f"{headers[0]:<{w1}}  {headers[1]:<{w2}}  {headers[2]:<{w3}}"
    print("\nКОНТРОЛЬ ПАРТІЇ:")
    print(line)
    print("-" * len(line))
    for r in results:
        print(f"{r[0]:<{w1}}  {r[1]:<{w2}}  {r[2]:<{w3}}")
    print()

# меню 
while True:
    print("Меню:")
    print("1. Додати препарат")
    print("2. Показати партію")
    print("3. Вийти")

    choice = input("Ваш вибір: ").strip()

    if choice == "1":
        name = input("Назва препарату: ").strip()
        qty_raw = input("Кількість: ").strip()
        try:
            qty = int(qty_raw)
        except:
            qty = qty_raw

        print("Категорія (antibiotic / vitamin / vaccine / ...)")
        cat = input("Введіть категорію: ").strip()

        print("Температура зберігання (оберіть діапазон):")
        print("a. нижче 5°C")
        print("b. від 5°C до 25°C")
        print("c. вище 25°C")
        t_choice = input("Ваш вибір: ").strip().lower()

        if t_choice == "a":
            temp = 2.0
        elif t_choice == "b":
            temp = 20.0
        elif t_choice == "c":
            temp = 30.0
        else:
            temp = "?"

        meds.append({"назва": name, "кількість": qty, "категорія": cat, "температура": temp})
        print(" Препарат додано!\n")

    elif choice == "2":
        print_meds(meds)

    elif choice == "3":
        print("Програму завершено.")
        break

    else:
        print("Невірний вибір, спробуйте ще раз.\n")
```
