```python
def is_number(x):
    return isinstance(x, (int, float)) and not isinstance(x, bool)

def category_by_amount(amount):
    if not is_number(amount) or amount < 0:
        return "Фальшиві дані"
    if amount < 100:
        return "Дрібнота"
    elif amount < 1000:
        return "Середнячок"
    else:
        return "Великий клієнт"

def decision_by_status(shortcut):
    s = (shortcut or "").strip().lower()
    match s:
        case "c":
            return "Працювати без питань"
        case "s":
            return "Перевірити документи"
        case "f":
            return "У чорний список"
        case _:
            return "Невідомий статус"

# Список клієнтів!
clients = []

def print_clients(clients):
    if not clients:
        print("\nСписок клієнтів порожній\n")
        return

    results = []
    for d in clients:
        name = d.get("ім'я", "")
        amount = d.get("сума угоди", "")
        status = d.get("статус перевірки", "")
        cat = category_by_amount(amount)
        dec = decision_by_status(status)
        results.append((name, amount, cat, dec))

    def fmt_amount(x):
        return f"{x:.2f}" if is_number(x) and x >= 0 else "—"

    rows = [(r[0], fmt_amount(r[1]), r[2], r[3]) for r in results]
    headers = ("Ім'я", "Сума", "Категорія", "Рішення")

    w1 = max(len(headers[0]), *(len(r[0]) for r in rows))
    w2 = max(len(headers[1]), *(len(r[1]) for r in rows))
    w3 = max(len(headers[2]), *(len(r[2]) for r in rows))
    w4 = max(len(headers[3]), *(len(r[3]) for r in rows))

    line = f"{headers[0]:<{w1}}  {headers[1]:>{w2}}  {headers[2]:<{w3}}  {headers[3]:<{w4}}"
    print("\nСПИСОК КЛІЄНТІВ:")
    print(line)
    print("-" * len(line))
    for r in rows:
        print(f"{r[0]:<{w1}}  {r[1]:>{w2}}  {r[2]:<{w3}}  {r[3]:<{w4}}")
    print()

# Головне меню 
while True:
    print("Меню:")
    print("1. Додати клієнта")
    print("2. Показати всіх клієнтів")
    print("3. Вийти")

    choice = input("Ваш вибір: ").strip()

    if choice == "1":
        name = input("Ім'я: ").strip()
        amount_raw = input("Сума угоди: ").strip()
        try:
            amount = float(amount_raw)
        except:
            amount = amount_raw
        print("Статус перевірки: (c = clean, s = suspicious, f = fraud)")
        status = input("Виберіть літеру: ").strip().lower()
        clients.append({"ім'я": name, "сума угоди": amount, "статус перевірки": status})
        print("Клієнта додано!\n")

    elif choice == "2":
        print_clients(clients)

    elif choice == "3":
        print("Програму завершено.")
        break

    else:
        print("Невірний вибір, спробуйте ще раз.\n")
```
